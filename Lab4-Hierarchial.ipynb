#### Import data from USCensus1990-2000rows.data.txt dataset

```python
import pandas as pd
from scipy.cluster.hierarchy import linkage,dendrogram
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

#### Remove the first column as it is not a feature

```python
data = df.iloc[:,1:]
data
```

#### Construct a Dendrogram

```python
Z = linkage(data,'ward')
plt.figure(figsize=(35,20))
dendrogram(Z)
plt.show()
```
